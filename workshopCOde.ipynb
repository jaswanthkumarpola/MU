```python
df.drop(["bike_name","city"],axis="columns",inplace=True)
```

```python
df.head()
#1 . x and y
# price ==> columns or dependent variable
 # x    === > kms_driven	owner	age	power	brand
```

```python
x = df[["kms_driven",	"owner",	"age",	"power",	"brand"]]
y = df[['price']]
```

```python
# train test split
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,train_size=0.8)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
```

```python
lnr = LinearRegression()  # machine 1
dtr  =  DecisionTreeRegressor()  #machine 2
rdf = RandomForestRegressor()  # machine 3
```

```python
x_train.shape
```

```python
### Train the machine or algorithms
lnr.fit(x_train,y_train)   # fit is used to train the machine or algorithm
dtr.fit(x_train,y_train)
rdf.fit(x_train,y_train)
print("successfully trained the algorithms!")
```

```python
# training score
print(lnr.score(x_train,y_train))
print(dtr.score(x_train,y_train))
print(rdf.score(x_train,y_train))
```

```python
# testing score
print(lnr.score(x_test,y_test))
print(dtr.score(x_test,y_test))
print(rdf.score(x_test,y_test))
```
